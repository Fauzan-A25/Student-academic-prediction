## Exploratory Data Analysis

```python
df = pd.read_csv('data.csv',sep=';')
```

```python
#Tipe data features

categorical = ['Marital status','Application mode','Course','Daytime/evening attendance\t','Previous qualification','Nacionality',
              "Mother's qualification","Father's qualification","Mother's occupation","Father's occupation","Displaced",
              "Educational special needs","Debtor","Tuition fees up to date","Gender","Scholarship holder","International","Target"]
numeric = [col for col in df.columns if col not in categorical]
```

```python
# fig,axs = plt.subplots(len(categorical),1,figsize=(12,8*len(categorical)))
# for i in range(len(categorical)):
#     sns.countplot(data=df,x=categorical[i],hue=categorical[i],ax=axs[i],palette='Set2',legend=False)
#     axs[i].set_title(f'Distribution of {categorical[i]}')
# plt.show()

fig,axs = plt.subplots(2,len(categorical)//2,figsize=(12*(len(categorical)//2),8*2))
j = 0
for i in range(len(categorical)):
    sns.countplot(data=df,x=categorical[i],hue=categorical[i],ax=axs[j,i%9],palette='Set2',legend=False)
    axs[j,i%9].set_title(f'Distribution of {categorical[i]}')
    j = (j + 1) % 2
plt.show()
```

```python
# Pairplot
plt.figure(figsize=(12*1.2,8*1.2))
corr_matrix = df[numeric].corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.show()
```

```python
def get_top_class(cat_feature,threshold=0.005):
    cp = cat_feature.value_counts()/len(df)
    cp = cp[cp >= threshold]
    return cp.index
```

```python
# Marital status
df['Marital status'] = df['Marital status'].apply(lambda x: x if x in [1,2,4,5] else 0)
# Application mode
df['Application mode'] = df['Application mode'].apply(lambda x: x if x in get_top_class(df['Application mode']) else -1)
# Previous qualification
df['Previous qualification'] = df['Previous qualification'].apply(lambda x: x if x in get_top_class(df['Previous qualification']) else 0)
# Mother's qualification
df["Mother's qualification"] = df["Mother's qualification"].apply(lambda x: x if x in get_top_class(df["Mother's qualification"]) else -1)
# Father's qualification
df["Father's qualification"] = df["Father's qualification"].apply(lambda x: x if x in get_top_class(df["Father's qualification"]) else -1)
# Mother's occupation
df["Mother's occupation"] = df["Mother's occupation"].apply(lambda x: x if x in get_top_class(df["Mother's occupation"]) else -1)
# Father's occupation
df["Father's occupation"] = df["Father's occupation"].apply(lambda x: x if x in get_top_class(df["Father's occupation"]) else -1)
```

```python
# fig,axs = plt.subplots(len(categorical),1,figsize=(12,8*len(categorical)))
# for i in range(len(categorical)):
#     sns.countplot(data=df,x=categorical[i],hue=categorical[i],ax=axs[i],palette='Set2',legend=False)
#     axs[i].set_title(f'Distribution of {categorical[i]}')
# plt.show()

fig,axs = plt.subplots(2,len(categorical)//2,figsize=(12*(len(categorical)//2),8*2))
j = 0
for i in range(len(categorical)):
    sns.countplot(data=df,x=categorical[i],hue=categorical[i],ax=axs[j,i%9],palette='Set2',legend=False)
    axs[j,i%9].set_title(f'Distribution of {categorical[i]}')
    j = (j + 1) % 2
plt.show()
```

```python
df.drop('Nacionality',axis=1,inplace=True)
```

```python
df['GDP'] = df['GDP']/(1+df['Inflation rate'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Target',axis=1),
                                                 df['Target'],
                                                 shuffle=True,
                                                 test_size=0.2,
                                                 random_state=42,
                                                 stratify=df['Target'])
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
categorical = ['Marital status','Application mode','Course','Daytime/evening attendance\t','Previous qualification',
              "Mother's qualification","Father's qualification","Mother's occupation","Father's occupation","Displaced",
              "Educational special needs","Debtor","Tuition fees up to date","Gender","Scholarship holder","International"]
train_pool = Pool(X_train,y_train,cat_features=categorical)
test_pool = Pool(X_test,y_test,cat_features=categorical)
```

```python
model = CatBoostClassifier(iterations=1000,loss_function='MultiClass',learning_rate=0.01)
model.fit(train_pool)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
yPred = model.predict(X_test)
report = classification_report(y_test,yPred)
```

```python
print(report)
```

```python
best_score = accuracy_score(y_test,yPred)
```

```python
useless_cols = ['Application order', 'Previous qualification', 'International']
```

```python
X_train.drop(useless_cols,axis=1,inplace=True)
X_test.drop(useless_cols,axis=1,inplace=True)
```

```python
categorical = [c for c in categorical if c not in useless_cols]
```

```python
model = CatBoostClassifier(iterations=1000,loss_function='MultiClass',learning_rate=0.01,verbose=False)
model.fit(X_train,y_train,cat_features=categorical)
```

```python
yPred = model.predict(X_test)
score = accuracy_score(y_test,yPred)
report = classification_report(y_test,yPred)
cfm = confusion_matrix(y_test,yPred,labels=model.classes_)
cfm_display = ConfusionMatrixDisplay(cfm,display_labels=model.classes_)
cfm_display.plot()
```

## Handle Imbalance Data

```python
model_b = CatBoostClassifier(iterations=1000,loss_function='MultiClass',learning_rate=0.01,verbose=False,auto_class_weights='SqrtBalanced')
model_b.fit(X_train,y_train,cat_features=categorical)
```

```python
yPred = model_b.predict(X_test)
score = accuracy_score(y_test,yPred)
report = classification_report(y_test,yPred)
cfm = confusion_matrix(y_test,yPred,labels=model.classes_)
cfm_display = ConfusionMatrixDisplay(cfm,display_labels=model.classes_)
cfm_display.plot()
```
