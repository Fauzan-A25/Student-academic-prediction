```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('data.csv',sep=';')
```

```python
df.shape
```

```python
df.head()
```

```python
df['Target'].unique()
```

```python
df['Target'] = df['Target'].map({'Dropout':0,'Graduate':1,'Enrolled':2})
```

```python
#Tipe data features

categorical = ['Marital status','Application mode','Course','Daytime/evening attendance\t','Previous qualification','Nacionality',
              "Mother's qualification","Father's qualification","Mother's occupation","Father's occupation","Displaced",
              "Educational special needs","Debtor","Tuition fees up to date","Gender","Scholarship holder","International"]
numeric = [col for col in df.columns if col not in categorical]
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Target',axis=1),df['Target'],shuffle=True,random_state=42,test_size=0.2)
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
train_pool = Pool(X_train,y_train,cat_features=categorical)
test_pool = Pool(X_test,y_test,cat_features=categorical)
```

```python
cb = CatBoostClassifier(iterations=1000,loss_function='MultiClass',verbose=True,learning_rate=0.01)
cb.fit(train_pool)
```

```python
pred = cb.predict(X_test)
```

```python
report_cb = classification_report(y_test,pred)
```

```python
print(report_cb)
```
